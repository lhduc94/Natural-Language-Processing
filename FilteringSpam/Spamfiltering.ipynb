```python
import os
import numpy as np
import pandas as pd
from collections import Counter
```

```python
def make_Dictionary(train_dir):
    emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]    
    all_words = []
    spam_words = []
    ham_words = []
    idx = -1
    for mail in emails:    
        with open(mail) as m:
            print (mail)
            idx += 1 
            for i,line in enumerate(m):
                if i == 2:  #Body of email is only 3rd line of text file
                    words = line.split()
                    all_words += words
                    if idx < 351 :
                        ham_words += words
                    else: spam_words += words
    
    dictionary = dict(Counter(all_words))
    spam_dictionary = dict(Counter(spam_words))
    ham_dictionary = dict(Counter(ham_words))
    
    df = pd.DataFrame([dictionary, ham_dictionary, spam_dictionary]).T
    df.columns = ['all', 'ham', 'spam']
    df = df.fillna(0)
    df['all'] += 2
    df['ham'] += 1
    df['spam'] += 1
    return df
```
