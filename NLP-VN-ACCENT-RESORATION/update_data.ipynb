```python
stopwords = []
import re
from nltk import ngrams
def transform_row(row,_stopwords=True):
    # Xóa số dòng ở đầu câu
    row = row.split("(")[0]
    row = re.sub(r"[0-9]", "", row)
    if _stopwords :
        for word in stopwords:
            row = row.replace(word,"")
    # Xóa dấu chấm, phẩy, hỏi ở cuối câu
    row = re.sub(r"[\.,\?]+$", "", row)
    # Xóa các kí tự khác
    char = ''',.;":'?!()-<>/'''
    for c in char:
        row = row.replace(c," ")
    row = row.replace(",", " ").replace(".", " ") \
        .replace(";", " ").replace("“", " ") \
        .replace(":", " ").replace("”", " ") \
        .replace('"', " ").replace("'", " ") \
#         .replace("!", " ").replace("?", " ") \
#         .replace("(", " ").replace(")", " ") \
#         .replace("-"," ").replace("div"," ") \
#         .replace("<"," ").replace(">"," ") \

    row = row.strip()
    return row 
from nltk import ngrams

def gen_ngram(string, n=1):
    gram_str = list(ngrams(string.split(), n))
    return [ " ".join(gram).lower() for gram in gram_str ]
```

```python
dictngram = {}
title = []
count = 0
with open('datasets/title.txt','r') as f:
    for line in f :
        title.append(line.strip().split('(')[0])
#         count+=1
print (len(title))
for name in title:
    for i in range(2,name.count(" ")+2):
        t = gen_ngram(name,i)
        for ngam in t : 
            ngam = ngam.lower().decode('utf-8')
            if dictngram.get(ngam,-1)==-1:
                dictngram[ngam] = 1
            else:
                dictngram[ngam] += 1
```

```python
accented_chars = {
    'a': u'a á à ả ã ạ â ấ ầ ẩ ẫ ậ ă ắ ằ ẳ ẵ ặ',
    'o': u'o ó ò ỏ õ ọ ô ố ồ ổ ỗ ộ ơ ớ ờ ở ỡ ợ',
    'e': u'e é è ẻ ẽ ẹ ê ế ề ể ễ ệ',
    'u': u'u ú ù ủ ũ ụ ư ứ ừ ử ữ ự',
    'i': u'i í ì ỉ ĩ ị',
    'y': u'y ý ỳ ỷ ỹ ỵ',
    'd': u'd đ',
}

plain_char_map = {}
for c, variants in accented_chars.items():
    for v in variants.split(' '):
        plain_char_map[v] = c


def remove_accent(text):
    return u''.join(plain_char_map.get(char, char) for char in text)

remove_accent(u'cô gái đến từ hôm qua')
```
