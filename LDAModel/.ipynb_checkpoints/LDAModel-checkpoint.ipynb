```python

import pandas as pd
df = pd.read_csv('dataNLP.csv', header=None,names=['title','description'])
f = open('stopwords.txt','r')
stopwords = ['tập)', '(phần','dù',"div"]
for word in f:
    stopwords.append(word.strip()+ ' ')
f.close()
df.head(20)
```

```python
import re

def transform_row(row):
    # Xóa số dòng ở đầu câu
    row = re.sub(r"[0-9]", "", row)
    for word in stopwords:
        row = row.replace(word,"")
    # Xóa dấu chấm, phẩy, hỏi ở cuối câu
    row = re.sub(r"[\.,\?]+$", "", row)
    
    # Xóa tất cả dấu chấm, phẩy, chấm phẩy, chấm thang, ... trong câu
#     row = row.replace(",", " ").replace(".", " ") \
#         .replace(";", " ").replace("“", " ") \
#         .replace(":", " ").replace("”", " ") \
#         .replace('"', " ").replace("'", " ") \
#         .replace("!", " ").replace("?", " ") \
#         .replace("(", " ").replace(")", " ") \
#         .replace("-"," ").replace("div"," ") \
#         .replace("<"," ").replace(">"," ") \

    row = row.strip()
    return row 

df['title'] = df.title.apply(transform_row)
df['description'] = df.description.apply(transform_row)
df.head(10)
```

```python
from nltk import ngrams
from pyvi.pyvi import ViTokenizer
def kieu_ngram(string, n=1):
    gram_str = list(ngrams(string.split(), n))
    return [ " ".join(gram).lower() for gram in gram_str ]
def str_tokenizer(string):
    gram_str = list(ViTokenizer.tokenize(string.decode('utf-8')).split())
    return gram_str
# df["title_1gram"] = df.title.apply(lambda t: kieu_ngram(t, 1))
# df["title_2gram"] = df.title.apply(lambda t: kieu_ngram(t, 2))
# df["description_1gram"] = df.description.apply(lambda t: kieu_ngram(t,1))
# df["description_2gram"] = df.description.apply(lambda t: kieu_ngram(t,2))
df["token_title"] = df.title.apply(lambda t : str_tokenizer(t))
df["token_description"] = df.description.apply(lambda t : str_tokenizer(t))
df.head(20)
```

```python
df["context"] = df["token_title"] + df["token_description"]
train_data = df.context.tolist()
```

```python
from gensim import corpora, models
import gensim
dictionary = corpora.Dictionary(train_data)
```

```python
corpus = [dictionary.doc2bow(text) for text in train_data]

# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=20, id2word = dictionary, passes=20)
```

```python
t = ldamodel.print_topics(num_topics=20, num_words=10)
for i,j in t:
    print i,j
```
